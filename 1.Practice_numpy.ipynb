## Two-dimensional Arrays (Tabular Representation of Data)

### Q-1) A company wishes to allocate the employee codes from 101 to 105 along with the ages 25, 26, 27, 28, and 29 (in years) to its employees. How can you represent this data using arrays?

```python
employee_ids = np.arange(101, 106)
ages = np.arange(25, 30)
```

### Columnar view of the data: To see the employee IDs along with the ages in a tabular orientation

```python
employee_data = np.column_stack((employee_ids, ages))
employee_data
```

### Q-3)  Updating 2D arrays:

#### 3.1: The employee with the employee code 103 wants to update the age to 31. How can you accomplish this task?

```python
employee_code = 103
row_index = np.where(employee_data[:, 0] == employee_code)[0][0] 
row_index
```

```python
employee_data[row_index, 1] = 31
employee_data
```

#### 3.2: The company wants to add the annual salaries of all the employees in the existing record. Add the annual salaries 35,000, 45,000, 40,000, 30,000, and 50,000 for the employees with the employee codes from 101 to 105, respectively.

```python
salaries = [35000, 45000, 40000, 30000, 50000] 
employee_data_with_salaries = np.column_stack((employee_data, np.array(salaries)))
employee_data_with_salaries
```

#### 3.3: The HR wants to update the salary of the employee with the employee code 102 to 43,500. Help the HR to accomplish this task.

```python
employee_code = 102
new_salary = 43500
row_index = np.where(employee_data_with_salaries[:, 0] == employee_code)[0][0]
employee_data_with_salaries[row_index, 2] = new_salary
employee_data_with_salaries
```

### Q-4) 2-D Arrays Manipulation:

The CEO of the organization has suggested to present all possible reports related to employees data to the HR team to understand the spread of salaries of the employees to take the decision for future recruitment and manage human resources expenses. How can you accomplish this task?


#### 4.1: Help Thomas, the financial analyst, wants to find the total annual salary of all employees.

```python
total_annual_salary = np.sum(employee_data_with_salaries[:, 2]) 
total_annual_salary 
```

#### 4.2: Help Thomas to calculate the average age of employees.

```python
average_age = np.mean(employee_data_with_salaries[:, 1])
average_age 
```

#### 4.3: The company has hired a new employee who is 27 years old, with an annual salary of 37,000 and the employee code 106. Add the employee details in the existing record of employees.

```python
new_employee = np.array([[106, 27, 37000]])
employee_data_with_salaries = np.vstack((employee_data_with_salaries, new_employee))
employee_data_with_salaries 
```

#### 4.4: The employee with the code 103 wishes to resign from the current job position. The company needs to remove the employee's data and update the record. How can you accomplish this task?

```python
employee_code_to_remove = 103
row_index = np.where(employee_data_with_salaries[:, 0] == employee_code_to_remove)
employee_data_with_salaries = np.delete(employee_data_with_salaries, row_index, axis=0)
employee_data_with_salaries 
```

#### 4.5: The company has decided to give a bonus amount to all the employees who are getting the annual salary more than $35,000. For this, the company needs to know the number of such employees along with their salaries.

```python
bonus_eligible_employees = emp_data_with_salaries[emp_data_with_salaries[:, 2] > 35000]
num_employees_with_bonus = len(bonus_eligible_employees)
num_employees_with_bonus 
```
